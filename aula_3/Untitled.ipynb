```python
# URL: http://www.anp.gov.br/images/infopreco/infopreco.csv
precos = pd.read_csv("infopreco.csv",
                     sep=';',
                     decimal=',',
                     parse_dates=[9],
                     dayfirst=True,
                     encoding='iso-8859-1')
```

```python
precos.head()
```

```python
precos.dtypes
```

```python
precos.groupby('UF')['MUNICÍPIO'].count().sum()
```

```python
precos.describe()
```

```python
t = pd.DataFrame(precos.groupby('UF')['MUNICÍPIO'])
```

```python
t.iloc[-3:]
```

```python
precos.groupby('UF').apply(lambda r: r['MUNICÍPIO'].iloc[0])
```

```python
precos.groupby('PRODUTO').apply(lambda r: precos.iloc[r['VALOR VENDA'].idxmax()])
```

```python
precos.groupby(['UF', 'MUNICÍPIO', 'PRODUTO'])['VALOR VENDA'].agg([min, max, len])
```

```python
def combustivel_basico(idx):
    return precos.loc[idx]['PRODUTO'].split(' ')[0]
```

```python
por_tipo = precos.groupby(combustivel_basico)
```

```python
por_tipo.agg(np.sum)
```

```python
por_tipo.size()
```

```python
por_tipo.agg([len, np.sum, np.mean, np.median])
```

```python
por_tipo['VALOR VENDA'].agg([len, np.sum, np.mean, np.median, np.min, np.max, np.std])
```

```python
por_tipo['VALOR VENDA'].agg({ "Amostras": len,
                              "Valor Médio": np.mean,
                              "Mediana": np.median,
                              "Valor Máximo": np.max})
```

```python
precos.groupby('UF').filter(lambda grupo: grupo.loc[(grupo['PRODUTO'] == 'Etanol'), 'VALOR VENDA'].mean() > 4)
```

Seleciona os preços de Etanol apenas dos municípios com mais de dois preços informados

```python
precos                                              \
   .loc[precos['PRODUTO'] == 'Etanol']              \
   .groupby(['MUNICÍPIO'])                          \
   .filter(lambda grupo: len(grupo) > 2)
```

```python
precos                                              \
   .loc[precos['PRODUTO'] == 'Etanol']              \
   .groupby(['MUNICÍPIO'])                          \
   .filter(lambda grupo: len(grupo) > 2, dropna=False)
```

`.head` e `.tail` são aplicados a cada um dos grupos!

```python
precos                                              \
   .loc[precos['PRODUTO'] == 'Etanol']              \
   .groupby(['MUNICÍPIO'])                          \
   .head(1) 
```
