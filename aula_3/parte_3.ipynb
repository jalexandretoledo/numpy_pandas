# GroupBy

```python
import numpy as np
import pandas as pd
```

```python
# fonte: http://dados.gov.br/dataset/1178-taxa-de-juros-selic-anualizada-base-252
df = pd.read_csv('bcdata.sgs.1178.csv',
                 sep=';',
                 decimal=',',
                 parse_dates=[0],
                 dayfirst=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['ano'] = df['data'].dt.year
```

```python
df['mes'] = df['data'].dt.month
```

```python
df['dia'] = df['data'].dt.day
```

```python
df.drop('data', inplace=True, axis='columns')
```

```python
df = df.loc[:, ['ano', 'mes', 'dia', 'valor']]
```

```python
df.head()
```

```python
df.groupby('ano')
```

```python
df.groupby('ano')['dia'].count()
```

```python
df.groupby('ano')['valor'].agg([min, max, np.mean, np.median, len])
```
